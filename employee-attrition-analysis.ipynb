<h3>Data Cleaning</h3>

```python
#Checking for duplicate values
print(data.duplicated().sum())
```

```python
#Converting all the column names to lower case
data.columns = data.columns.str.lower()

print(data.columns)
```

```python
# Convert categories to numbers
data["attrition"] = data['attrition'].astype('category').cat.codes

data.describe()
```

```python
#Checking details of stock option level column
print(data['stockoptionlevel'].describe())

#Checking for outliers in stock option level column
plt.boxplot(data["stockoptionlevel"])
plt.show()
```

```python
#As we see we have outliers so we will remove them as they are low in number and removing them does'nt effect our results.
data = data[data["stockoptionlevel"] < 2.5 ]

#Checking for outliers again
plt.boxplot(data["stockoptionlevel"])
plt.show()
```

```python
#Checking details of stock option level column
print(data['totalworkingyears'].describe())

#Checking for outliers in stock option level column
plt.boxplot(data["totalworkingyears"])
plt.show()
```

```python
data = data[data["totalworkingyears"] < 25.5 ]

#Checking for outliers again
plt.boxplot(data["totalworkingyears"])
plt.show()
```

```python
#Checking details of training times last year column
print(data['trainingtimeslastyear'].describe())

#Checking for outliers in training times last year
plt.boxplot(data["trainingtimeslastyear"])
plt.show()
```

```python
data = data[(data['trainingtimeslastyear'] > 0.5) & (data['trainingtimeslastyear'] < 4.5)]

#Checking for outliers again
plt.boxplot(data["trainingtimeslastyear"])
plt.show()
```

```python
#Checking for outliers in years at company
plt.boxplot(data["yearsatcompany"])
plt.show()
```

```python
#As we see we have outliers so we will remove them as they are low in number and removing them does'nt effect our results.
data = data[data["yearsatcompany"] < 18 ]

#Checking for outliers again
plt.boxplot(data["yearsatcompany"])
plt.show()
```
